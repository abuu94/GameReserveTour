 Game Reserve Tour:
 
 This is my first game  where you need to choose the right answer on animals with four legs to get free ticket to visit one among best Tanzania National Game Reserve.

```python
#I have import two modules/libraries

import random   
import time


# the first tuples contain mixture of animals with different number of legs.
animal=('cow','goat','girrafe','lion','tiger','hippo','hen','duck','fish','parrot','kangaroo')

# the first tuples contain mixture of animals with four legs only
four_leg_animal=('cow','goat','girrafe','lion','tiger','hippo')

#For the improvement of the game , turples can be filled with many items


#the below function provide introduction and instruction of the game.
def message():
    print('You are in a Serengeti Reserve Park. Located in Tanzania Country,')
    print('Your are give a Test. Simple Test of Mentioning Four Legs Animal')
    print('If You pass the Test , You will get free Tour to Visit Game Reserve')
    print('Be Carefull , Some Animal may be not comman to you.')
    print()
    
# the below function is a main , it perform core functionality of the game

def start():
    n=1         # this variable n  is for initial value of while loop
    guess=0     # this variable  is for counting correct answer that exist in four_leg_animals tuple
    guess1=0    # this variable  is for counting correct answer that exist out of four_leg_animals tuple
    
    
    
    print("How many Question You want in the Test : ")    #Prompt user to number of question to be asked
    q=input()                                             #Takes user input for no. of question to be asked
    
    
    
    while n<=int(q):                    #while loop starts from n (fisrt question) upto q (last question)
        print(n,'Question :')
       

        inp=random.choice(animal)        #random question generated in each iteration
        
        
        
        print("Animal ",inp," has 4 legs..  Enter True or False :")
        user=input()
        # 2 lines above Prompt user a question and take his/her respond.
        
        
        #the for loop below check if answer is correct by checking whether random number selected, user respond, and 
        #four_leg_animals satisfy condition
        for x in four_leg_animal:
            if (inp == x) and (user=="True"):
                guess=guess+1

                
        #the if statement below check if answer is correct by checking user respond that doesnot validate with the
        #tuple, and four_leg_animals satisfy condition
        if user=="False" and inp not in four_leg_animal:
            guess1=guess1+1 
            
            
        n+=1   #this code update the while condition
    print("*********************************************")
    print("Please Wait.. for result*")
  
    time.sleep(2)      #this code is for making some delay of the program, for good interactivity
    print("*********************************************")
    time.sleep(2)      #this code is for making some delay of the program, for good interactivity
    
    
    result= guess + guess1  #this code is summing answers
    print("Correct :",result ," Out of ",q," question")
    time.sleep(2)
    
    #if condition below does checking if all answer are correct or not
    #if all are correct , means player has passed the test
    #if any  is incorrect , means player has failed the test
    if int(result) ==int(q):
        print("Congratulation ,!! You have passed the Test. Enjoy the Tour")
    else:
        print("Sory ,!! You have failed . You have to answer all Question Correctly to get the Tour")
        print("\n")
        


play = 'yes'   #this is sentiment for allowing the gain to be played


#Also the below block of code does enable a player to play again and again
while play == 'yes' or play == 'y':

    message()     #this line  call message function which provide instructions
    start()       #this line  call start function for start playing

    print('Do you want to play again? (yes or no)')      #this line ask player for playing again

    play = input()
    
# end of  Game Reserve Tour

```
